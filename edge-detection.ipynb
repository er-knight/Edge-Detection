```python
BASE_URL = "https://raw.githubusercontent.com/er-knight/edge-detection/main"
```

```python
from skimage import io

image = io.imread(BASE_URL + "/butterfly.jpg")
```

```python
import matplotlib.pyplot as plt

plt.imshow(image);
```

```python
import cv2

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image
```

```python
canny = cv2.Canny(image, threshold1=180, threshold2=200)
plt.imshow(canny);
```

```python
cv2.imwrite("edges.jpg", canny)
```

```python
# apply gaussian blur to reduce noise

blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
plt.imshow(blurred_image);
```

```python
canny = cv2.Canny(blurred_image, threshold1=180, threshold2=200)
plt.imshow(canny);
```

```python
cv2.imwrite("edges-blurred.jpg", canny)
```

```python
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale_image);
```

```python
blurred_image = cv2.GaussianBlur(grayscale_image, (5, 5), 0)
plt.imshow(blurred_image);
```

```python
canny = cv2.Canny(blurred_image, threshold1=180, threshold2=200)
plt.imshow(canny);
```

```python
cv2.imwrite("edges-grayscale-and-blurred.jpg", canny)
```
